```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

# Pre fine-tuning

```python
# sampled from the same dataset
with open('finetuning_results/results_pre_ft.json', mode='r') as f:
    results_pre_ft = json.load(f)

# gpt prompt-res pairs
with open('finetuning_results/results_pre_ft_essential.json', mode='r') as f:
    results_pre_ft_essential = json.load(f)

```

# Post fine-tuning

```python
# sampled from the same dataset
with open('finetuning_results/results_post_ft.json', mode='r') as f:
    results_post_ft = json.load(f)

# gpt prompt-res pairs
with open('finetuning_results/results_post_ft_essential.json', mode='r') as f:
    results_post_ft_essential = json.load(f)
```

# Plot

```python
labels = [prompt['prompt'] for prompt in results_pre_ft]
pre_ft_X = [prompt['log_prob'] for prompt in results_pre_ft]
post_ft_X = [prompt['log_prob'] for prompt in results_post_ft]

x = np.arange(len(labels))  # positions for bars
width = 0.35  # bar width

fig, ax = plt.subplots()

# Bars side by side
rects1 = ax.bar(x - width/2, pre_ft_X, width, label="pre fine-tuning")
rects2 = ax.bar(x + width/2, post_ft_X, width, label="post fine-tuning")

ax.set_ylabel("Log Probs")
ax.set_xticks(x)
ax.legend()

plt.show()
```
